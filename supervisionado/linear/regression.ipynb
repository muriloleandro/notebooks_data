```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Gerar dataset
X = np.linspace(0, 4*np.pi, 50)
Y = []
for x in X:
    epsilon = np.random.normal(0, 0.4)
    y = x + 3*np.sin(x) + epsilon
    Y.append(y)
```

```python
plt.scatter(X, Y)
```

```python
# Base polinomial
phi = []
phi.append(lambda x: 1)
for n in range(9):
    phi.append(lambda x, n=n: x**n)

# Design Matrix
Phi = np.zeros((len(X), len(phi)))
for n in range(len(X)):
    for j in range(len(phi)):
        Phi[n][j] = phi[j](X[n])

Phi_inv = np.linalg.pinv(Phi)
w = np.matmul(Phi_inv, Y)
```

```python
def f(x):
    phi_x = [phi[j](x) for j in range(len(phi))]
    return np.inner(w, phi_x)

plt.scatter(X, Y)
plt.plot(X, [f(x) for x in X])
plt.show()
```
